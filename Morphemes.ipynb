Вычленяем морфемы из текста грамматики (зпатем вручную убираем некоторые)

```python
with open("2_monografiya_sovremennyy_mansi_yaz_2017_-_maket_4.txt", encoding="utf-8") as f:
    text = f.read()
```

```python
symbols = {" ", ",", ".", "\n", ":"}
for i in range(10):
    symbols.add(str(i))
```

Суффиксы

```python
c = 0
suffixes = []
while c < len(text):
    if text[c] == " " and text[c + 1] == "-" and text[c + 2] != " ":
        c1 = c + 1
        while text[c1] not in symbols:
            c1 += 1
        suffixes.append(text[c:c1])
        c = c1
    else:
        c += 1
```

```python
set(suffixes)
```

```python
suffixes_set_initial = set(suffixes)
len(suffixes_set_initial)
```

```python
suffixes_set =set()
for el in suffixes_set_initial:
    if "/" in el:
        suffixes_set.add(el[:el.find("/")])
        suffixes_set.add(el[el.find("/") + 1:])
    elif "(" in el and ")" in el and el.count("-") == 1:
        suffixes_set.add(el[:el.find("(")])
        suffixes_set.add(el.replace("(", "").replace(")", ""))
    else:
        suffixes_set.add(el)
```

```python
suffixes_set.remove("")
for el in list(suffixes_set):
    try:
        if "+" in el:
            suffixes_set.remove(el)
            suffixes_set.add(el.replace("+-", "").replace("+", ""))
    except:
        pass
```

```python
for el in list(suffixes_set):
    try:
        if "(" not in el and el[-1] == ")":
            suffixes_set.remove(el)
            suffixes_set.add(el[:-1])
        while el[0] == " ":
            suffixes_set.remove(el)
            suffixes_set.add(el[1:])
        
    except:
        pass
```

```python
for el in list(suffixes_set):
    try:
        while el[0] == " ":
            suffixes_set.remove(el)
            suffixes_set.add(el[1:])
        
    except:
        pass
```

```python
suffixes_set
```

```python
for el in list(suffixes_set):
    try:
        if "(" in el:
            suffixes_set.remove(el)
            suffixes_set.add(el[:el.find("(")])
            suffixes_set.add(el[el.find("(")+1:el.rfind(")")])
        
    except:
        pass
```

```python
len(suffixes_set)
```

```python
import json

with open("suffixes.json", "w", encoding="utf-8") as f:
    json.dump(list(suffixes_set), f, indent=4, ensure_ascii=False)
```

Префиксы

```python
c = 0
prefixes = []
while c < len(text):
    if text[c] in symbols and text[c - 1] == "-":
        c1 = c - 2
        while text[c1] not in {" ", "(", "\n"}:
            c1 -= 1
        prefixes.append(text[c1:c])
    c += 1
```

```python
prefixes = {el for el in prefixes if len(set(el).intersection(set(map(str, range(10))))) == 0}
```

```python
set(prefixes)
```

```python
prefixes_set = set(prefixes)
for el in list(prefixes_set):
    try:
        while el[0] == " ":
            prefixes_set.remove(el)
            prefixes_set.add(el[1:])
        while el[0] == "\n":
            prefixes_set.remove(el)
            prefixes_set.add(el[1:])
        while el[0] == "(":
            prefixes_set.remove(el)
            prefixes_set.add(el[1:])
    except:
        pass
```

```python
prefixes_set
```

```python
import json

with open("prefixes.json", "w", encoding="utf-8") as f:
    json.dump(list(prefixes_set), f, indent=4, ensure_ascii=False)
```

Интерфиксы

```python
c = 0
interfixes = []
while c < len(text):
    if text[c] == "-":
        c1 = c + 1
        while c1 - c < 5:
            c1 += 1
            if text[c1] == "-" and len(set(text[c:c1+1]).intersection(symbols|{"+", "("})) == 0:
                interfixes.append(text[c:c1+1])
        c = c1
    else:
        c += 1
```

```python
set(interfixes)
```

```python
interfixes_set = set(interfixes)
```

```python
import json

with open("interfixes.json", "w", encoding="utf-8") as f:
    json.dump(list(interfixes_set), f, indent=4, ensure_ascii=False)
```
