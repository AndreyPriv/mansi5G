Выделение суффиксов в словах из словаря

```python
import json

with open("Afanasieva_Sobyanina_dictionary.json", encoding="utf-8") as f:
    words_dict = json.load(f)
```

```python
with open("suffixes.json", encoding="utf-8") as f:
    suffixes = json.load(f)
suffixes_set = set([el[1:] for el in suffixes])
```

```python
suffixes_set
```

```python
len(words_dict)
```

```python
set_from_dict = set(words_dict.keys())
```

```python
for word in set_from_dict:
    print(word, words_dict[word].replace("\n", ""))
    while(len(word) > 1):
        for i in range(len(word) - 1, 0, -1):
            if word[-i:] in suffixes_set:
                print(word, "=", word[:-i], "+", word[-i:])
                break
        else:
            break
        word = word[:-i]
    else:
        print(word)
    print()
```

Сопоставление словаря с датасетом

```python
import pandas as pd
```

```python
df = pd.read_csv("overall_80K.csv", index_col=0)
df
```

```python
def delete_punct(s):
    for k in [".", ",", "!", "?", "\""]:
        s = s.replace(k, "")
    return s

words = []
for el in df["target"]:
    words += delete_punct(el).split()
len(words)
```

```python
len(set(words))
```

```python
words_set = set(words)
```

Сколько словоформ из словаря встречается в датасете:

```python
len({el for el in words_dict.keys() if el in words_set})
```
