```python
with open("2_monografiya_sovremennyy_mansi_yaz_2017_-_maket_4.txt", encoding="utf-8") as f:
    text = f.read()
```

```python
text
```

```python
symbols = {",", ".", ";", ":", "’", "–"}
```

```python
c = 0
phrases = []
translations = []
while c < len(text):
    if text[c] == "‘":
        c1 = c
        while c1 < len(text):
            c1 += 1
            if text[c1] == "’":
                
                c2 = c
                while c > 0:
                    if text[c2] in symbols or text[c2-3:c2+1] == " от ":
                        break
                    else:
                        c2 -= 1
                if c - c2 > 2:
                    phrase = text[c2+1:c].replace("\n", " ").replace("-", "")
                    if "(" in phrase and ")" in phrase:
                        phrase = phrase[:phrase.find("(")] + phrase[phrase.rfind(")") + 1:]
                    if "[" in phrase and "]" in phrase:
                        phrase = phrase[:phrase.find("[")] + phrase[phrase.rfind("]") + 1:]
                    if phrase[0] == " ":
                        phrase = phrase[1:]
                    if phrase[-1] == " ":
                        phrase = phrase[:-1]
                    translation = text[c+1:c1].replace("\n", " ").replace("-", "")
                    phrases.append(phrase)
                    translations.append(translation)
                c = c1
                break      
    else:
        c += 1
    
```

```python
import pandas as pd


df = pd.DataFrame({"phrase": phrases, "translation": translations})
df
```

```python
df.to_csv("phrases_from_grammar.csv")
```
